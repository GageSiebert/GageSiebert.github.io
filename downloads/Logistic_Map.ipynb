```python
import numpy as np
import matplotlib.pyplot as plt


it=50000
ittwo=500
read=ittwo-1
counts=[]
fixedpoints=[]
history=np.zeros(ittwo)
a=np.linspace(3.4,4,num=it)

for n in range(1,it): #for every 'a' value
    x=0.5
    for step in range(0,ittwo): #iterates function to find pattern
        x=a[n]*x*(1-x) 
        history[step]=x
        
    for m in range(1,round(ittwo/4)):
        
        if history[read]<history[read-m]+0.0001 and history[read]>history[read-m]-0.0001:
        
            for c in range(1,m):                   
                counts.append(a[n])
                fixedpoints.append(history[m-c])
            break  
        
    
#print(counts)
plt.scatter(counts,fixedpoints,s=0.1)
fig = plt.gcf()
fig.set_size_inches(11,8)
```
